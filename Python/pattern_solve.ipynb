```python
#square hollow pattern
for i in range(1,7,1):
    for j in range(1,7,1):
        if i==1 or i==6 or j==1 or j==6:
            print("*", end =" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#number triangular
for i in range(1,5,1):
    
    for j in range(5,i,-1):
        print(" ",end="")
    for j in range(1,i+1,1):
        print(i, end=" ")
       
    print()
```

```python
#number increasing pyramid
for i in range(1,5,1):
    for j in range(1,i+1,1):
        print(j, end=" ")
    print()
```

```python
#number increasing reverse pyramid
for i in range(6,1,-1):
    for j in range(1,i-1,1):
        print(j, end=" ")
    print()
```

```python
#number changing pyramid
num=1
for i in range(1,5,1):
    for j in range(1,i+1,1):
        print(num,end=" ")
        num+=1
    print()
```

```python
#zero-one triangle
for i in range(1,5,1):
    for j in range(1,i+1,1):
        if ((i+j)%2)==0:
            print("1", end=" ")
        else:
            print("0",end=" ")
    print()
```

```python
#palindrome triangular

for i in range(1,5,1):

    for j in range(5,i,-1):
        print(" ",end=" ")
        
    for j in range(i,0,-1):
        print(j,end=" ")


    for j in range(2,i+1):
        print(j,end=" ")

    print()

    
```

```python
#rhombus pattern
for i in range(1,6,1):
    for j in range(1,i,1):
        print(" ",end= "")
        
    for j in range(1,5,1):
        print("*",end=" ")
    print()
    
```

```python
#dimond pattern
for i in range(1,5,1):
    for j in range(5,i,-1):
        print(" ",end="")
    for j in range(1,i+1,1):
        print("*",end =" ")
    print()

for i in range(1,4,1):
    for j in range(0,i+1,1):
        print(" ",end="")
    
    for j in range(4,i,-1):
        print("*",end=" ")
    print()

```

```python
#butterfly star pattern
n=4
for i in range(1,5,1):
    for j in range(i):
        print("*",end=" ")
   
    for j in range(2*(n-i)):
        print(" ",end=" ")

    for j in range(i):
        print("*",end=" ")
    print()

for i in range(4,0,-1):
    for j in range(i):
        print("*",end=" ")
    for j in range(2*(n-i)):
        print(" ",end=" ")
    
    for j in range(i):
        print("*",end=" ")
    
    print()



```

```python
#hollow triangle pattern 
for i in range(1,6,1):
    for j in range(5,i,-1):
        print(" ",end="")
    for j in range(1,6,1):
        if i==5 or j==1 or j==i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
#pascal's triangle

for i in range(0,4):
    for j in range(4-i-1):
        print(" ",end="")
    num=1
    for j in range(i+1):
        print(num,end=" ")
        num=num*(i-j)//(j+1)
    print()

```
