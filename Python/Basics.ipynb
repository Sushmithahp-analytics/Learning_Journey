```python
#if else if chain/if else if ladder/continueous if else
"""take percentage imput from user 
85 to 100 print distinction
60 to 84 print 1st class
50 to 59 print 2nd class
35 to 49 print pass
<35 print fail"""

marks=float(input("enter the marks: "))
if (marks<0 or marks>100):
    print("invalid input")
elif (marks>=85):
    print("Distinction")
elif (marks >=60):
    print("First class")
elif (marks >=50):
    print("Second class")
elif (marks >=35):
    print("pass")
else:
   print("fail")
```

```python
#write a program to check the given input forms which triangle
#1.equilateral
#2.isoceles
#3.scalene
#4.invalid triangle

a= float(input("enter side a value:"))
b= float(input("enter side b value:"))
c= float(input("enter side c value:"))

if((a+b)<c or (a+c)<b or (b+c)<a):
    print("invalid triangle")
elif (a==b==c):
    print("it is equilateral triangle")
elif ((a==b) or(a==c) or (b==c)):
    print("it is isoceles triangle")
else:
    print("it is scalene triangle")
```

```python
#calculate final bill amount
#purchase amount of 0 to 5000 no discount
#purchase amount of 5000 to 10000 is 10%discount
#purchase amount of 10000 to 20000 is 15% discount
#purchase amount is 20000 to 30000 is 20 % discount
#purchase amount is above 30000 flat 30% discount

p_amount= int(input("enter purchase amount:"))
if (p_amount>0 and p_amount<5000):
    print("no discount")
    b_amount=p_amount
elif p_amount>=5000 and p_amount<=10000:
    print("10 % discount")
    discount=p_amount*0.1
    b_amount=p_amount-discount
elif p_amount>=10000 and p_amount<=20000:
    print("15% discount")
    discount=p_amount*0.15
    b_amount=p_amount-discount
elif p_amount>=20000 and p_amount<=30000:
    print("20% discount")
    discount=p_amount*0.2
    b_amount= p_amount-discount
elif p_amount>30000:
    print("flat 30% discount")
    discount=p_amount*0.3
    b_amount= p_amount-discount
else:
    print("invalid amount")
    discount = "invalid"
    b_amount = "invalid"
print(f"purchase amount is:{p_amount}\ndiscount anount :{discount}\ntotal bill amount {b_amount}")
```
