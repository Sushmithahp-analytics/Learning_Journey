```python
#perform sum of digits using while loop

num = int(input("enter a number:")) #initialization
sum = 0
while(num>0):
    rem = num%10
    sum += rem
    num = num//10 #decrementation
print("the sum of digit is :",sum)
```

```python
#check the number is palindrome or not
num = int(input("enter a number:")) #initialization
org = num
rev = 0
while(num>0):
    rem = num%10
    rev = (rev * 10 ) +rem
    num = num//10 #decrementation
print(f"original value: {org} reversed value: {rev}" )
if (org==rev):
    print(f"{org} is palindrome")
else:
    print(f"{org} is not palindrome")
```

```python
c=65 #ASCII value of upper letter starts from 65 and lower starts from 97
i=1
while (i<=5): 
    #calculates the space
    for j in range(5,i,-1): 
        print(" ",end="")
    #star print
    for k in range (0,i,1): 
        print(chr(c), end =" ")
        c+=1
    print()
    i+=1
```
