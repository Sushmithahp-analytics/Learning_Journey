```python
#function without input and without return

def prime():
    num1 = int(input("enter start number:"))
    num2 = int(input("enter stop number:"))
    for i in range(num1,num2+1,1):
        for j in range(2,i,1):
            if (i%j==0):
                break
        else:
            print(i,end=" ")

prime()
```

```python
#with input and without return

def prime1(s1,s2):
    for i in range(s1,s2+1):
        if (i>1):
            for j in range(2,i,1):
                if (i%j==0):
                    break
            else:
                print(i,end=" ")

prime1(,20)
                
```

```python
#without input with return

def prime2():
    pl= []
    num1 = int(input("enter start number:"))
    num2 = int(input("enter stop number:"))
    for i in range(num1,num2+1,1):
        if (i>1):
            for j in range(2,i,1):
                if (i%j==0):
                    break
            else:
                pl.append(i)
    return pl

num = prime2()
print(num)
    
```

```python
#with input and with return

def prime3(num1,num2):
    pl =[]
    for i in range(num1,num2+1):
        if (i>1):
            for j in range(2,i,1):
                if (i%j==0):
                    break
            else:
                pl.append(i)
    return pl

prime3(2,20)
```

```python
#create function to calculate factorial of number without loops
#recursive function (function call itself)
def factorial(num):
    if num==0 or num==1:
        return 1
    else:
        return num*factorial(num-1)

factorial(5)  
```

```python
#create function to calculate sum of number of given range without loop

def sum(stv,spv):
    if stv == spv:
        return stv
    else:
        return stv+(sum(stv+1,spv))

sum(1,20)
        
```

```python
#function calling another function
def add1(a,b):
    return a+b
def sqadd(x,y):
    sqadd=add1(x,y)**2
    return sqadd
sqadd(2,5)
```

```python
#nested function
def greeting(name,time):
    def msg(time):
        if time==1:
            return "morning"
        elif time ==2:
            return "afternoon"
        else:
            return"evening"
    print(f"hello {name},good {msg(time)}")

greeting("priyanka",3)
```
